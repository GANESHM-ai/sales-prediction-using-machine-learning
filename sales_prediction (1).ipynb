```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder

from google.colab import files
files.upload('')
sp = pd.read_csv('sales_prediction.csv')

le = LabelEncoder()
sp['Store_Type'] = le.fit_transform(sp['Store_Type'])

x = sp[['Store_Type','Promotion','Holiday']]
y = sp['Sales']

model = LinearRegression()
model.fit(x,y)


new_data_point = pd.DataFrame([[1, 0, 0]]) #columns = ['Store_Type', 'Promotion', 'Holiday'])
prediction = model.predict(new_data_point)
prediction


print(f'Predicted sales for new patient: {prediction[0]}')

```
